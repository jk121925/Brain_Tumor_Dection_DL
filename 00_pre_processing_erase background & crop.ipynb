```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_origin = cv2.imread('C:\\Users\\jk121\\Desktop\\temp\\archive\\Testing\\glioma_tumor\\image(10).jpg')
```

```python
img = cv2.cvtColor(img_origin,cv2.COLOR_BGR2RGB)
out_img = img
plt.imshow(out_img)
```

```python
lower = np.array([30,30,30])
higher = np.array([250,250,250])
```

```python
mask = cv2.inRange(out_img,lower,higher)
```

```python
plt.imshow(mask,'gray')
```

```python
cont,_ = cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
```

```python
cont_img = cv2.drawContours(out_img,cont,-1,255,2)
```

```python
plt.imshow(cont_img)
```

```python
c = max(cont,key =cv2.contourArea)
```

```python
x,y,w,h = cv2.boundingRect(c)
```

```python
cv2.rectangle(out_img,(x,y),(x+w,y+h),(0,255,0),3)
plt.imshow(out_img)
```

```python
crop_img = img_origin[y: y+h, x : x+w]
plt.imshow(crop_img)
cv2.imshow("image",crop_img)
cv2.waitKey(0)
```

```python
cv2.imwrite('temp1.jpg',crop_img)
```
